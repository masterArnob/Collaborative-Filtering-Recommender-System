```python
rating_w_books = ratings.merge(books, on="isbn")
rating_w_books.head()
```

```python
grouped_books = rating_w_books.groupby("book_name")
grouped_books
```

```python
total_ratings = grouped_books["ratings"].count()
total_ratings = total_ratings.reset_index()
total_ratings.rename({
    "ratings": "total_ratings"
    }, axis=1, inplace=True)
total_ratings.head()
```

```python
avg_ratings = grouped_books["ratings"].mean()
avg_ratings = avg_ratings.reset_index()
avg_ratings.rename({
    "ratings": "avg_ratings"
    }, axis=1, inplace=True)
avg_ratings.head()
```

```python
df = total_ratings.merge(avg_ratings, on="book_name")
df.head()
```

```python
df["total_ratings"] > 250
```

```python
df[df["total_ratings"] > 250]
```

```python
df[df["total_ratings"] > 250].sort_values(by="avg_ratings", ascending=False)
```

```python
famous_books = df[df["total_ratings"] > 250].sort_values(by="avg_ratings", ascending=False).head(50)
famous_books
```

```python
famous_books.merge(books, on="book_name")
```
