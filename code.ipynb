```python
def recommend(book_name):
    # fetch index of the book_name
    book_index = np.where(pt.index == book_name)[0][0]
    simmilar_items = sorted(list(enumerate(similarity[book_index])), key=lambda x: x[1], reverse=True)[1:6]
    
    recommendations = []
    
    for item in simmilar_items:
        recommendations.append(pt.index[item[0]])
    return recommendations
```

```python
recommend("The Notebook")
```

```python
model_data = {
    "data": books,
    "similarity": similarity,
     "pivot_index": pt.index.tolist()  # <-- save only books used for similarity
}
```

```python
import joblib
joblib.dump(model_data, "model.pkl")
print("Model Saved Successfully!")
```

```python
loaded_model = joblib.load("model.pkl")
loaded_model_prediction = recommend("Toxin")
print(f"Prediction from loaded model: {loaded_model_prediction}")

for book in loaded_model_prediction:
    print(book)
```
